```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Titanic.csv")
df
```

```python
sns.boxplot(data=df,x='sex',y='age',hue='survived')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Distribution of Age by Gender and Survival')
```

This code will create a box plot with the x-axis representing the "sex" column, the y-axis representing the "age" column, and different box plots will be displayed based on the "survived" column. The "survived" column is used as a hue to differentiate the boxes based on survival status
